```python
import networkx as nx
```

```python
G = nx.DiGraph() #Grafica dirigida, segun nuestros parametros
#Creamos las interacciones entre los puntos y sus pesos. Vertices y aristas
G.add_edge("O","A",weight = 2.0)
G.add_edge("O","B",weight = 5.0)
G.add_edge("O","C",weight = 4.0)
G.add_edge("A","D",weight = 7.0)
G.add_edge("A","B",weight = 2.0)
G.add_edge("B","C",weight = 1.0)
G.add_edge("B","E",weight = 3.0)
G.add_edge("B","D",weight = 4.0)
G.add_edge("D","E",weight = 1.0)
G.add_edge("D","T",weight = 5.0)
G.add_edge("E","T",weight = 7.0)
```

```python
#Graficamos el grafo o mapa dadas las anteriores cordenadas.
grafico_base = G.subgraph([i for i in G])
nx.draw(grafico_base, with_labels=True)
```

```python
#Definimos la ruta que queremos identifocar como mas corta
inicio = "O"
fin = "T"
#USamos el algoritmo dijkstra para ubicar dicha ruta
ruta_optima= nx.dijkstra_path(G, source = inicio, target = fin, weight = True )
#Consultamos el numero de pasos o estaciones.
pasos = nx.dijkstra_path_length(G, source = inicio, target = fin, weight = True)
print(f"La ruta mas optima es {ruta_optima} y el numero de vertices o estaciones es {pasos}")
```

```python
#Graficamos la ruta dada como resultado
grafico = G.subgraph([i for i in ruta_optima])
nx.draw(grafico, with_labels=True)
```
